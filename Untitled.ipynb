```python
#Find interquartilic range per state
res =df_state_filter.groupby("State")["Perpetrator Age"].quantile([0.25,0.75]).unstack(level = 1)
res['IQR'] = res[0.75]-res[0.25] 
res_IQR= res['IQR']
pd.DataFrame(res_IQR)#IQR calculation for each Drug Regimen
```

```python
#Total Number of datapoints per state, and outliars
res['Lower_Q-1.5IQR']= res[0.25]-1.5*res['IQR']
res['Higher_Q+1.5IQR']=res[0.75]+1.5*res['IQR']


df_state_filter_outliars= df_state_filter.loc[((res.loc[df_state_filter['State'], 'Lower_Q-1.5IQR'] > df_state_filter["Perpetrator Age"].values) | (df_state_filter["Perpetrator Age"].values > res.loc[df_state_filter['State'],'Higher_Q+1.5IQR'])).values]
df_state_filter_outliars =df_state_filter_outliars.groupby(['State'])['Perpetrator Age'].count()
df_state_filter_outliars = pd.DataFrame(df_state_filter_outliars)
df_state_filter_outliars.rename(columns= {'Perpetrator Age': 'Amount of Potential Outliars'})


df_state_filter_datacount = pd.DataFrame(df_state_filter.groupby('State')['Perpetrator Age'].count())


df_state_filter_outliars['Total Datapoints'] = df_state_filter_datacount
df_state_filter_outliars['Percentage of outliars'] =(df_state_filter_outliars['Perpetrator Age'] /df_state_filter_outliars['Total Datapoints']) *100
df_state_filter_outliars
```

```python
bar_chart = df_state_filter_outliars.plot(kind='bar',y='Percentage of outliars', title = 'Percentage of outliars',figsize=(20,5))
bar_chart.legend().remove()
bar_chart.set_xlabel('State')
bar_chart.set_ylabel('Percentage of Outliars')
bar_chart
```
